```python
#Reading data from API
import pandas as pd
import requests
```

```python
response=requests.get('https://www.freetestapi.com/api/v1/movies') #pass api link to get function of requests
```

```python
response.json()
```

```python
df = pd.DataFrame(response.json()).head()[['id','year','rating']] #Using pandas Dataframe 
```
