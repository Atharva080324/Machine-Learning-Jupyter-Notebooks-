2) Opening a csv file from url 

```python
import pandas as pd
import requests
from io import StringIO
url = ""
headers={""}
req = requests.get(url,headers=headers)
data = StringIO(req.text)
pd.read_csv(data)
```

```python
def rename(name):
    if name=="Royal Challengers Bangalore":
       return 'RCB'
    else :
        return name
    if name=="Kolkata Knight Riders":
       return 'KKR'
    else :
        return name
```

```python
#12)Convertors specific name into desired names
#this changed Royal Challengers Bangalore to RCB in column team1 with function rename 
pd.read_csv('IPL MATCHES 2008-2020.csv',converters={'team1':rename,'team2':rename}).head()
```
