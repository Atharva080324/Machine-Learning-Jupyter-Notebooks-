To solve the line equation we use two approach:

1)Closed form solution (OLS): in which the formula is already defined 

2)Non Closed form : In which there are more number of dimensions. Solved using gradient
descent.


Internally in sklearn OLS if implemented 



1)OLS(Ordinary least squares):


y = mx+b

m = slope of the regression line

b = intercept (y-axis intercept)

$$m = \frac{\sum_{i=1}^{n} (x_i - \bar{x})(y_i - \bar{y})}{\sum_{i=1}^{n} (x_i - \bar{x})^2}$$


$$b = \bar{y} - m \bar{x}$$


```python
class CustomLR:
    def __init__(self):
        self.m = None
        self.b = None

    def fit(self,X_Train,Y_Train):
        num,den = 0,0
        
        for i in range(X_Train.shape[0]):
            num += (X_Train[i] - X_Train.mean()) * (Y_Train[i] - Y_Train.mean()) #calculating numerator 
            den += (X_Train[i]- X_Train.mean())*(X_Train[i]- X_Train.mean())
        self.m = num/den
        self.b = Y_Train.mean() - (self.m*X_Train.mean())
        print(self.m)
        print(self.b)
    def predict(self,X_Test):
        return self.m * X_Test + self.b

    
```
