```python
date['order_date'] = pd.to_datetime(date['order_date'])      #convert object dtype to datetime dtype
```

```python
date['order_year'] = date['order_date'].dt.year               #extracting year 
```

```python
date.head()
```

```python
date['order_month_no'] = date['order_date'].dt.month                    #month number
```

```python
date.head()
```

```python
date['order_month'] = date['order_date'].dt.month_name()             #month name
```

```python
date.head()
```

```python
date['day'] = date['order_date'].dt.day                                        #extract day of month
```

```python
date.sample(5)
```

```python
date['day_name'] = date['order_date'].dt.day_name()                              #extract dayname of month 
```

```python
date.sample(5)
```

```python
date['date_is_weekend'] = np.where(date['day_name'].isin(['Sunday','Saturday']),1,0)    #give 1 where day is sat or sun else 0
date = date.drop(columns={'ship_mode','customer_id','sales'})
date.sample(5)
```

```python
date['week_of_year'] = date['order_date'].dt.isocalendar().week
```
