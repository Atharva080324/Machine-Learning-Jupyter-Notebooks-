1. Missing Values Imputation
What it is: Sometimes, datasets have missing values, which can cause issues in model training. Imputation helps fill in these gaps.

Example:
Imagine a customer database where some people have not provided their age. Instead of leaving these empty, we can fill them with the average age of all customers or use the most common age in the dataset.

2. Handling Categorical Variables
What it is: Machine learning models work with numbers, so we need to convert non-numerical (categorical) data into numerical form.

Example:
Suppose a dataset contains the colors of cars ("Red," "Blue," "Green"). A model can't process these as words, so we replace them with numbers like:

Red → 1

Blue → 2

Green → 3

3. Outlier Detection
What it is: Outliers are extreme values that can mislead a model. Detecting and handling them improves accuracy.

Example:
A company records employee salaries. Most employees earn between ₹30,000 and ₹1,00,000, but one record shows ₹10,00,000. This could be a mistake or an unusual case, and we may need to remove or adjust it.

4. Feature Scaling
What it is: Ensuring all numerical features are in a similar range to prevent bias in model training.

Example:
If a dataset has "Height (in cm)" ranging from 150 to 200 and "Weight (in kg)" ranging from 50 to 100, the difference in scale might affect predictions. Scaling both to a range of 0-1 ensures fair treatment.

5. Feature Construction
What it is: Creating new, useful features from existing ones to improve model performance.

Example:
An e-commerce website tracks the purchase date of users. From this, we can create new features like:

Day of the week (Monday, Tuesday, etc.)

Holiday or not (Weekday vs. Weekend)

Time of purchase (Morning, Afternoon, Night)

6. Feature Selection
What it is: Choosing the most relevant features while removing unnecessary ones to make the model more efficient.

Example:
A real estate price prediction model may have features like:

Number of bedrooms

House color

Distance from school

Floor type
Here, "House color" might not impact the price much, so we can remove it to simplify the model.

7. Feature Extraction
What it is: Converting complex data into a more useful and compact form.

Example:

A text analysis system extracts key words from long documents instead of using entire text.

A facial recognition system extracts important features like the distance between eyes rather than storing full images.
