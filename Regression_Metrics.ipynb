```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('placement.csv')
```

```python
df.head()
```

```python
plt.scatter(df['cgpa'],df['package'])
plt.xlabel('CGPA')
plt.ylabel('Package(in LPA)')
```

```python
X = df.iloc[:,0:1]
Y = df.iloc[:,-1]
```

```python
X,Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,Y_train)
```

```python
plt.scatter(df['cgpa'],df['package'])
plt.plot(X_train,lr.predict(X_train),color='red')
plt.xlabel('CGPA')
plt.ylabel('Package in LPA')
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
Y_pred = lr.predict(X_test)
```

```python
Y_test.values
```

```python
print("Mean Absolute Error",mean_absolute_error(Y_test,Y_pred)) #passing actual values and predicted values
```

```python
print("Mean Squared Error",mean_squared_error(Y_test,Y_pred)) #passing actual values and predicted values
#answer not in lpa
```

```python
print("Reduced Mean Absolute Error",np.sqrt(mean_squared_error(Y_test,Y_pred))) #passing actual values and predicted values)
```

```python
print('R2 Score',r2_score(Y_test,Y_pred))
```
